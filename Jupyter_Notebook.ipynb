### The Code in cell 5 is suppose to run the cars st data set and show a simple graph of the Horsepower and Weight, with a colored points to show the transmission.

```r
library(datasets)
data(mtcars)

library(ggplot2)

mtcars$am <- as.factor(mtcars$am)
cols <- c("#1170AA", "#55AD89")
ggplot(aes(x=wt,y=hp,color=am),data=mtcars)+geom_point()+ggtitle("HorsePower VS Weight (1000 lbs)")+labs(x="Weight per 1000lbs",y="Horsepower",color="Transmission")+scale_color_discrete(labels = c("Automatic", "Manual"))
```
